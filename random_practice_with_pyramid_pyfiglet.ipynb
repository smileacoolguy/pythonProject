```python
n=10
for i in range(n):
    for j in range(1,i+1):
        print(i*i,end=" ")
    print()
print("print x program ")
for i in range(1,n):
    for j in range(1,n):
        if i==j or j==n-i:
            print(i,end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
p=[]
np=[]
for f in range(100):
    if f <=1:
        np.append(f)
    elif f==2:
        p.append(f)
    elif f%2!=0:
        p.append(f)
    else:
        np.append(f)
print("Prime"+str(p))
print("NonPrime"+str(np))
    
```
